# Tutorial 1. Fine-grained data retrieval for stimulus selection

This tutorial demonstrates how to retrieve a targeted subset of lexical items from the released familiarity datasets for stimulus design.

In this example, we use the GPT-4o word-level dataset to extract the top 10% most familiar two-character items.

**Working path**
- Notebook: `Tutorials/01_stimulus_selection.ipynb`
- Input file: `../Data/2 norms/norms_gpt4o_word_prompt.xlsx`
- Output file: `./output/gpt4o_top10pct_two_character_words.xlsx`

**Execution note**
This tutorial notebook is intended for local execution after the repository has been downloaded.

## Step 1. Import packages and define file paths

```python
import math
from pathlib import Path
import pandas as pd

# -----------------------------
# File paths
# -----------------------------
input_path = Path("../Data/2 norms/norms_gpt4o_word_prompt.xlsx")
output_dir = Path("./output")
output_dir.mkdir(parents=True, exist_ok=True)

output_path = output_dir / "gpt4o_top10pct_two_character_words.xlsx"
```

## Step 2. Load the released familiarity dataset

```python
df = pd.read_excel(input_path)

print(f"Loaded {len(df)} rows.")
df.head()
```

## Step 3. Check the required columns

This tutorial uses the following variables:
- `WORD`: the lexical item
- `Length`: the number of characters in the item
- `GPT_FAM_probs`: the GPT-4o familiarity estimate

Detailed definitions of released variables are provided in `../Data/Variable Descriptions.xlsx`.

```python
required_cols = ["WORD", "Length", "GPT_FAM_probs"]
missing_cols = [col for col in required_cols if col not in df.columns]

if missing_cols:
    raise ValueError(f"Missing required columns: {missing_cols}")

print("All required columns are available.")
print(df[required_cols].head())
```

## Step 4. Restrict the data to two-character items

```python
df_two = df[df["Length"] == 2].copy()

print(f"Number of two-character items: {len(df_two)}")
df_two.head()
```

## Step 5. Sort items by familiarity and select the top 10%

```python
df_two = df_two.sort_values(by="GPT_FAM_probs", ascending=False)

n_select = math.ceil(len(df_two) * 0.10)
df_top10 = df_two.head(n_select).copy()

print(f"Selected top {n_select} items.")
df_top10.head()
```

## Step 6. Export the selected subset

```python
df_top10.to_excel(output_path, index=False)

print(f"Saved output to: {output_path}")
```

## Notes

- The same workflow can be applied to other released datasets by changing the input file and the familiarity column of interest.
- For the Qwen-max dataset, replace `GPT_FAM_probs` with `qwen_FAM_mean_30`.
- The output file generated here can be used directly for later stimulus screening and subset selection.
